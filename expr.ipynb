```python
import sympy as sp
import random
sp.init_printing(use_unicode=False, wrap_line=True)
```

```python
from sympy import symbols, expand, latex
x, y, a, b = symbols('x, y, a, b')
```

```python
'''
**********************************************
*               LaTeX変換部                   *
**********************************************
'''
def ltx():
    fhalf, shalf = expr()
    return (
        r"\noindent" "\n"
        r"\begin{minipage}[t]{0.46\textwidth}" "\n"
        r"\begin{enumerate}[label=\arabic*.,leftmargin=8mm,itemsep=25mm,topsep=20mm]" "\n"
        rf"{"\n".join(fhalf)}" "\n"
        r"\end{enumerate}" "\n"
        r"\end{minipage}%" "\n"
        r"\hfill\vrule\hfill"
        r"\begin{minipage}[t]{0.46\textwidth}" "\n"
        r"\begin{enumerate}[start = 8,label=\arabic*.,leftmargin=8mm,itemsep=25mm,topsep=20mm]" "\n"
        rf"{"\n".join(shalf)}" "\n"
        r"\end{enumerate}" "\n"
        r"\end{minipage}%" "\n"
    )
```

```python
'''
**********************************************
*              問題作成部１                    *
**********************************************
'''
def expr():
    problems = []
    #14問1セットで生成しておこう.
    for i in range(14):
        while True:
    #a,bは-9から9で生成しておく．
            a = random.randint(-9,9)
            b = random.randint(-9,9)
    #(x + 3)(x + 3)とか(x + 0)(x + 2)はあったらよくない！
            if a == 0 or b == 0 or a == b: 
                continue
            problem_latex = latex(sp.expand((x + a) * (x + b)))
            problems.append(rf"\item ${problem_latex}$")
            break
    fhalf = problems[:7]
    shalf = problems[7:]
    return (fhalf, shalf)
```

```python
for i in range(1,6): #名前に関しては0から始まるのは気持ち悪いですね...
    with open(f"output{i}.tex", "w") as f:
        f.write(ltx())
    print(f"ファイル{i}を作成しました")
```
